# Parsing m_id
there appears to be 2 variations in representation.
`[%02x]A7`
and
`D1[%04X]A7`

```python
def m_id_parse(string):
    m = re.match('^d1([0-9a-f]{4})a7$', string)
    if m == None:
        m = re.match('^([0-9a-f]{2})a7$', string)
        if m == None:
            # this should never happen.
            print(string)
            return
    return int(m.group(1), 16)

m_id_parse('3ba7')
```

```python
dt['m_id'] = dt['m_id'].apply(m_id_parse)
dt.head()
```

# Parsing m_Label
It seems the pattern here is either `"DA [%04Length of String] [String] A8"` or `"A[%01length of string] [String] A8"`

```python
def m_Label_parse(string):
    m = re.match('^da[0-9a-f]{4}([0-9a-f]+)a8$', string)
    if m == None:
        m = re.match('^a[0-9a-f]{1}([0-9a-f]+)a8$', string)
        if m == None:
            print(string)
            return
    
    return bytearray.fromhex(m.group(1)).decode()

m_Label_parse('da0014466f726573745f456e656d795f3031615f303031a8')
```

```python
dt['m_Label'] = dt['m_Label'].apply(m_Label_parse)
dt.head()
```

# Parsing MaxHP
There appears to be 3 patterns: `[%02Hex]A7`, `D1[%04Hex]A7`, `D2[%08Hex]A7`

All of which appear to be necessary due to insufficient space to store integers.

```python
def m_maxhp_parse(string):
    m = re.match('^([0-9a-f]{2})a7$', string)
    if m == None:
        m = re.match('^d1([0-9a-f]{4})a7$', string)
        if m == None:
            m = re.match('^d2([0-9a-f]{8})a7$', string)
            if m==None:
                print(string)
                return
    return int(m.group(1), 16)
m_maxhp_parse('d200046cd0a7')
```

```python
dt['m_MaxHP'] = dt['m_MaxHP'].apply(m_maxhp_parse)
dt.head()
```

# Parsing m_MaxSP
This appears to follow the same pattern as m_MaxHP, with the exception: the terminator is a6.

```python
def m_maxsp_parse(string):
    m = re.match('^([0-9a-f]{2})a6$', string)
    if m == None:
        m = re.match('^d1([0-9a-f]{4})a6$', string)
        if m == None:
            m = re.match('^d2([0-9a-f]{8})a6$', string)
            if m==None:
                print(string)
                return
    return int(m.group(1), 16)
m_maxsp_parse('d100aaa6')
```

```python
dt['m_MaxSP'] = dt['m_MaxSP'].apply(m_maxsp_parse)
dt.head()
```

Interestingly, m_AtkP, m_DefP, m_AtkM, m_DefM, m_Agi, m_Crt, m_CrtDef all seem to follow the same principle, with different ending bytes.

```python
def m_general_int_parse(string):
    m = re.match('^([0-9a-f]{2})a[0-9a-f]$', string)
    if m == None:
        m = re.match('^d1([0-9a-f]{4})a[0-9a-f]$', string)
        if m == None:
            m = re.match('^d2([0-9a-f]{8})a[0-9a-f]$', string)
            if m==None:
                print(string)
                return
    return int(m.group(1), 16)
```

```python
for key in ['m_AtkP', 'm_DefP', 'm_AtkM', 'm_DefM', 'm_Agi', 'm_Crt', 'm_CrtDef']:
    dt[key] = dt[key].apply(m_general_int_parse)
dt.head()
```

# m_EquipAtk parsing
Annoyingly, this is represented in a completely different format than any other variable thus far. This appears to always have a `da0011` 3-byte ending hex. I **do not** know what this represents, and any theories are welcome.

The structure appears to be similar, if we ignore the ending hex discreptancy. To reiterate, the structure is `[%02Hex]DA0011`, `D1[%04Hex]DA0011`, and `D2[%08Hex]DA0011`.

```python
def m_equipatk_parse(string):
    m = re.match('^([0-9a-f]{2})da0011$', string)
    if m == None:
        m = re.match('^d1([0-9a-f]{4})da0011$', string)
        if m == None:
            m = re.match('^d2([0-9a-f]{8})da0011$', string)
            if m==None:
                print(string)
                return
    return int(m.group(1), 16)
m_equipatk_parse('d100a6da0011')
```

```python
dt['m_EquipAtk'] = dt['m_EquipAtk'].apply(m_equipatk_parse)
dt.head()
```

# Parsing m_DropReward, m_DisplaySpecialSkillGauge, m_GenerateSpecialSkillValueOrverTurn, m_ReduceSpecialSkillValueBreak

Much like m_EquipAtk, these stats appear to follow the "3-byte long ending marker" rule. m_MaxSpecialSkillValue appears to have a 4-byte long pattern.

m_DropReward's structure: `[%02Hex]da0012`, then `d1[%04Hex]da0012` if out of space.

m_DisplaySpecialSkillGauge structure: `[%02Hex]da0024`

m_GenerateSpecialSkillValueOrverTurn structure: `[%02Hex]da001e`

m_ReduceSpecialSkillValueBreak structure: `[%02Hex]da0016`

m_MaxSpecialSkillValue structure: `[%02Hex]de0023a4`

```python
def m_dropreward_parse(string):
    m = re.match('^([0-9a-f]{2})da0012$', string)
    if m == None:
        m = re.match('^d1([0-9a-f]{4})da0012$', string)
        if m == None:
            print(string)
            return
    return int(m.group(1), 16)
m_dropreward_parse('01da0012')
```

```python
dt['m_DropReward'] = dt['m_DropReward'].apply(m_dropreward_parse)
```

```python
def m_DisplaySpecialSkillGauge_parse(string):
    m = re.match('^([0-9a-f]{2})da0024$', string)
    if m == None:
        print(string)
        return
    return int(m.group(1), 16)
m_DisplaySpecialSkillGauge_parse('c2da0024')
```

```python
dt['m_DisplaySpecialSkillGauge'] = dt['m_DisplaySpecialSkillGauge'].apply(m_DisplaySpecialSkillGauge_parse)
```

```python
def m_GenerateSpecialSkillValueOrverTurn_parse(string):
    m = re.match('^([0-9a-f]{2})da001e$', string)
    if m == None:
        print(string)
        return
    return int(m.group(1), 16)
dt['m_GenerateSpecialSkillValueOrverTurn'] = dt['m_GenerateSpecialSkillValueOrverTurn'].apply(m_GenerateSpecialSkillValueOrverTurn_parse)
```

```python
def m_ReduceSpecialSkillValueBreak_parse(string):
    m = re.match('^([0-9a-f]{2})da0016$', string)
    if m == None:
        print(string)
        return
    return int(m.group(1), 16)
dt['m_ReduceSpecialSkillValueBreak'] = dt['m_ReduceSpecialSkillValueBreak'].apply(m_ReduceSpecialSkillValueBreak_parse)
```

```python
def m_MaxSpecialSkillValue_parse(string):
    m = re.match('^([0-9a-f]{2})de0023a4$', string)
    if m == None:
        print(string)
        return
    return int(m.group(1), 16)
dt['m_MaxSpecialSkillValue'] = dt['m_MaxSpecialSkillValue'].apply(m_MaxSpecialSkillValue_parse)
```
